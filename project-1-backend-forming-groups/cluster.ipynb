Create a list of possible feature vectors. Since features have weights, it is best to assign them values -1 and 1, so that features with -1 and 1 are placed as far apart as possible. Multiply them with weights so we can form clusters considering those weights. 

```python
import itertools
lst = list(itertools.product([-1, 1], repeat=8))
wt_lst = [100000, 1000, 100, 10, 1000, 10, 100, 1000]
lst_of_vectors = list()


for feature_vector in lst:
    a_vector = list()
    for x, elem in enumerate(feature_vector):
        a_vector.append(elem * (wt_lst[x]))
    lst_of_vectors.append(a_vector)
print lst_of_vectors
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array(lst_of_vectors)
X
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
kmeans.labels_
```

If we have people with *all* different choices and there are only two clusters possible, they should be put in different clusters. Let's verify. 

```python
kmeans.predict([[-100000, -1000, -100, -10, -1000, -10, -100, -1000]])
```

```python
kmeans.predict([[100000, 1000, 100, 10, 1000, 10, 100, 1000]])
```

```python
kmeans.cluster_centers_
```
